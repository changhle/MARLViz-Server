```python
import csv
import requests
import os
```

# Fetch Data

```python
file_path = 'data/230217/230217.csv'

record_list = []

with open(file_path, "r", encoding='UTF-8') as csv_file:
    csvreader = csv.reader(csv_file, delimiter=",", quotechar='"')
    next(csvreader)
    for row in csvreader:
        record_id = row[0]
        user_id = row[1]
        start_time = row[2]
        equip_id = row[5]
        count = int(row[6])
        duration = float(row[7])
        data_url = row[8]
        
        item = {
            'id': record_id,
            'user': user_id,
            'start': start_time,
            'equip': equip_id,
            'count': count,
            'duration': duration,
            'url': data_url
        }
        
        record_list.append(item)
        
        file_name = '{}.txt'.format(record_id)
        
        
        txt_path = 'data/230217/record_files/{}'.format(file_name)
        print(txt_path)
        
        response = requests.get(data_url)
        
        with open(txt_path, "w", encoding='UTF-8') as text_file:
            text_file.write(response.text)
```
